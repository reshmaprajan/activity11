```python
# Separate the data into feature data and target data (X and y, respectively)
```

```python
x=data.drop(['passed'],axis=1)
y=pd.DataFrame(data['passed'])
```

# Preprocessing Feature Columns

```python
# label encoding
```

```python
from sklearn import preprocessing
lb_en = preprocessing.LabelEncoder()
z=data.drop(['age', 'Medu', 'Fedu','Mjob', 'Fjob', 'reason', 'traveltime', 'studytime',
       'failures','famrel', 'freetime', 'goout', 'Dalc',
       'Walc', 'health', 'absences', 'passed'],axis=1)
for i in z:
    z[i]=lb_en.fit_transform(z[i])
z=pd.DataFrame(z)
z.columns
```

```python
x=x.drop(['school', 'sex', 'address', 'famsize', 'Pstatus', 'guardian','nursery',
       'schoolsup', 'famsup', 'paid', 'activities', 'higher', 'internet',
       'romantic'],axis=1)
x=pd.concat([z,x],axis=1)
x.columns
```

```python
x.head()
```

```python
# one hot encoding
```

```python
x = pd.get_dummies(x)
x.head()
```

# Implementing, Training and Testing Data Split

```python
# splitting the data into train and test
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=0.25)
```

# Training and Evaluating Models

```python
# Importing the supervised learning models from sklearn
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
```

```python
print('Accuracy_score is',round(accuracy_score(y_test,y_pred),2))
```
